```python
import requests
import json
```

```python
# Source: Search for the TOKEN
```

```python
header = {'User-Agent' : ('Mozilla/5.0 (X11; Linux x86_64) '
                          'AppleWebKit/537.36 (KHTML, like Gecko) ' 
                          'Chrome/85.0.4183.83 Safari/537.36'),
         'Content-type': 'Application/json',
         }


s = requests.Session()
s.headers = header
```

```python
URL = 'https://flightaware.com/live/flight/DLH772'
```

```python
r = s.get(URL)
```

```python
get_token = json.loads(r.text[r.text.find('trackpollGlobals')+19:r.text.find('USERTOKEN')+23])
```

```python
token = get_token.get('TOKEN')
```

```python
token
```

```python
TRACKER = 'https://flightaware.com/ajax/trackpoll.rvt'
params = dict(token=token,
              locale='en_US', 
              summary=1)
```

```python
g = requests.get(TRACKER, params=params)
```

```python
g.reason
```

```python
g.json()
```

```python
MAIN = 'https://flightaware.com/live/'
r = s.get(MAIN)
```

```python
start = r.text.find('mapGlobals') + 13
end = r.text[start:].find('</script>') - 1
```

```python
response = json.loads(r.text[start:start+end])
response
```

```python
token = response.get('VICINITY_TOKEN')
token
```

```python
MAIN_API = 'https://flightaware.com/ajax/vicinity_aircraft.rvt'
params = dict(minLon=-180,
          minLat=31.95648193359375,
          maxLon=-102.28792905807495,
          maxLat=36.9140625,
          token=token
         )
g = s.get(MAIN_API, params=params)
g
```
