```python
from collections import defaultdict
import pandas as pd
from flights import flightaware
```

```python
flight = flightaware.Flight() 
flight.get(icao_id='DLH772')  
```

```python
flights = flightaware.Flights()
flights.get()
```

```python
data = flights.get()
```

```python
data['features'][0]
```

```python
data['features'][5]['properties']
```

```python

special_properties = ['altitude', 'altitudeChange','groundspeed']

storage = defaultdict(list)
for feature in data['features']:
    storage['latitude'].append(feature['geometry']['coordinates'][0])
    storage['longitude'].append(feature['geometry']['coordinates'][1])
    
    [storage[key].append(value) 
     for key, value in feature['properties'].items()
     if key not in special_properties]
    
    [storage[special].append(feature['properties'].get(special, None))
                            for special in special_properties]

    
```

```python
pd.DataFrame(storage)
```
